```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://quotes.toscrape.com/'
response = requests.get(url)
trainings_soup =  BeautifulSoup(response.text, 'html.parser')
```

```python
trainings_soup
```

```python
fin_names = []
names = trainings_soup.findAll('span',attrs={"class":"text"})
for trainings_name in names:
    fin_names.append(trainings_name.text)
for i in range(0, len(fin_names)):
    fin_names[i] = fin_names[i].strip('“”')
fin_names
```

```python
all_quotes, all_authors, all_links, all_tags = [], [], [], []

for product in trainings_soup.find_all('div',attrs={"quote"}):
    print(product.find('a').get('href'))
    link = 'https://quotes.toscrape.com/' + product.find('a').get('href')
    quote = product.find('span').text
    author = product.find('small', class_='author').text
    tags = [tag.text for tag in product.find_all('a', class_='tag')]
    #r2 = requests.get(link)
    #c = r2.content
    #soup2=BeautifulSoup(c,"html.parser")
    #author = soup2.find('small', class_ = "author").text  
    #tags = soup2.find('a', class_ = 'tag').text

    all_tags.append(tags)
    all_links.append(link)
    all_authors.append(author)
    all_quotes.append(quote)


data_list = list(zip(all_quotes,all_authors, all_links, all_tags))
column_names = ['Quotes','Authors', 'Links','Tags']
qss = pd.DataFrame(data_list,columns=column_names)
```
