```python
session = HTMLSession()

response1 = session.get('https://bina.az/')

soup = BeautifulSoup(response1.text, 'html.parser')

print(soup)
```

```python
all_locations, all_date_time, all_price = [], [], []

selectors = [
    #'.residences_list',
    #'.items_slider_row--vipped',
    '.items-i',
    '.residences-i',
    #'.items-i featured vipped',
    #'.items-i vipped'
]

products = soup.select(', '.join(selectors))

for product in products:
    location_tag = product.find('div', class_='location')
    location = location_tag.text.strip() if location_tag else None
    
    date_time_tag = product.find('div', class_=['city_when', 'deadline'])
    date_time = date_time_tag.text.strip() if date_time_tag else None
    
    price_tag = product.find_all('span', class_='price-val')
    price = price_tag[0].text.strip() if price_tag else None

    all_locations.append(location)
    all_date_time.append(date_time)
    all_price.append(price)
```

```python
links = []
for k in soup.find_all('div', attrs={'class':'residences_list'}):
    a_tags = k.find_all('a')
    for a_tag in a_tags:
        href = a_tag.get('href')
        if href:
            full_link = 'https://bina.az' + href
            links.append(full_link)
links
```

```python
items = soup.find_all('div', class_='items-i')

# Iterate through each item and find the <a> tag within it
for item in items:
    link_tag = item.find('a', class_='item_link')
    if link_tag and link_tag.get('href'):
        href = link_tag.get('href')
        full_link = 'https://bina.az' + href
        links.append(full_link)
```

```python
links
```

```python
data_list = list(zip(all_locations, all_date_time, all_price, links))
column_names = ['Locations', 'Date_time','Price', 'Links']
qss = pd.DataFrame(data_list,columns=column_names)
```
